```python
import numpy as np
```

```python
u = np.arange(16).reshape(4,4) #0~15までの数の4*4の行列
u
```

```python
np.roll(u, 1, axis = 0) #下からはみ出たものが上にくるのか
```

```python
np.roll(u, -1, axis = 0) #上にはみ出たものが下にくる
```

```python
np.roll(u, 1, axis = 1) #上記と異なる回転軸で(右に一つの回転)
```

```python
np.roll(u, -1, axis = 1)
```
